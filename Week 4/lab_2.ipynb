```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('marketing_customer_analysis_clean.csv')  
df
```

```python
df.select_dtypes(include=np.number)
```

# X-y SPLIT

```python
X = df[['customer_lifetime_value', "income", "monthly_premium_auto", "months_since_last_claim", "months_since_policy_inception", "number_of_open_complaints", "number_of_policies"]]
y = df[['total_claim_amount']]
```

# Train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 100)
```

# SCALING

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
#We just have to apply fit once (the X_train)
```

```python
scaler.fit(X_train)
```

```python
scaler.fit(y_train)
```

```python
#Now we transform BOTH: train and test
```

```python
scaler.fit(y_test)
```

```python
scaler.fit(X_test)
```

# Creating our model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
model = lm.fit(scaler.transform(X_train), y_train)
```

```python
model.coef_
```

```python
model.intercept_
```
