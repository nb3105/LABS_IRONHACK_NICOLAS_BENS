```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import random
```

```python
url = "https://digitaldreamdoor.com/pages/best_songs-Alternative.html"
response = requests.get(url)
response.status_code
soup = BeautifulSoup(response.content, "html.parser")
```

```python
songs = soup.select("div.list")
```

```python
artists = soup.select("div.list span:nth-child(n+1)")
artists = [(i,j) for i,j in enumerate(artists) if i%2!=0]
```

```python
df = pd.DataFrame(list(zip(songs, artists)))
```
